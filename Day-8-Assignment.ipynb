```python
import tensorflow as tf
from tensorflow.keras.datasets import reuters
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, LSTM, GRU, Dense
import matplotlib.pyplot as plt
```

```python

num_words = 15000
maxlen = 250  

(x_train, y_train), (x_test, y_test) = reuters.load_data(num_words=num_words)

x_train = pad_sequences(x_train, maxlen=maxlen)
x_test = pad_sequences(x_test, maxlen=maxlen)

num_classes = max(y_train) + 1
y_train = to_categorical(y_train, num_classes)
y_test = to_categorical(y_test, num_classes)
```

# **RNN**

```python
model = Sequential([
    Embedding(input_dim=num_words, output_dim=32, input_length=maxlen),
    SimpleRNN(32),  
    Dense(num_classes, activation='softmax') 
])
```

```python
model.compile(optimizer='adam',
              loss='categorical_crossentropy',  # Because labels are one-hot
              metrics=['accuracy'])

model.summary()
```

```python
history = model.fit(
    x_train, y_train,
    epochs=20,
    batch_size=128,
    validation_split=0.2
)
```

```python
test_loss, test_acc = model.evaluate(x_test, y_test)
print(f"Test Accuracy: {test_acc:.4f}")
```

```python
# Accuracy plot
plt.plot(history.history['accuracy'], label='Train Acc')
plt.plot(history.history['val_accuracy'], label='Val Acc')
plt.title('Accuracy over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()
```

```python

# Loss plot
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title('Loss over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
```

# **LSTM**

```python
model_lstm = Sequential([
    Embedding(input_dim=num_words, output_dim=32, input_length=maxlen),
    LSTM(32),
        Dense(num_classes, activation='softmax') 

])

model_lstm.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model_lstm.summary()
```

```python
history_lstm = model_lstm.fit(
    x_train, y_train,
    epochs=20,
    batch_size=128,
    validation_split=0.2
)
```

```python
test_loss, test_acc = model_lstm.evaluate(x_test, y_test)
print(f"LSTM Test Accuracy: {test_acc:.4f}")
```

```python
# Accuracy
plt.plot(history_lstm.history['accuracy'], label='Train Acc')
plt.plot(history_lstm.history['val_accuracy'], label='Val Acc')
plt.title('LSTM: Accuracy over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()

# Loss
plt.plot(history_lstm.history['loss'], label='Train Loss')
plt.plot(history_lstm.history['val_loss'], label='Val Loss')
plt.title('LSTM: Loss over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
```

# **GRU**


```python
model_gru = Sequential([
    Embedding(input_dim=num_words, output_dim=32, input_length=maxlen),
    GRU(32),
    Dense(num_classes, activation='softmax') 

])

model_gru.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model_gru.summary()
```

```python
history_gru = model_gru.fit(
    x_train, y_train,
    epochs=20,
    batch_size=128,
    validation_split=0.2
)
```

```python
test_loss, test_acc = model_gru.evaluate(x_test, y_test)
print(f"GRU Test Accuracy: {test_acc:.4f}")
```

```python
# Accuracy
plt.plot(history_gru.history['accuracy'], label='Train Acc')
plt.plot(history_gru.history['val_accuracy'], label='Val Acc')
plt.title('GRU: Accuracy over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()

# Loss
plt.plot(history_gru.history['loss'], label='Train Loss')
plt.plot(history_gru.history['val_loss'], label='Val Loss')
plt.title('GRU: Loss over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
```

##  Model Evaluation and Observations

We evaluated three different recurrent neural network architectures: **Vanilla RNN**, **GRU**, and **LSTM**, using the same dataset and training parameters. Below are the results and comparative observations.

###  Test Accuracy Results:
- **Vanilla RNN**: 43.68%
- **GRU**: 53.29%
- **LSTM**: 56.54%

---

###  Observations:

1. **LSTM Outperformed Other Models**  
   - LSTM achieved the highest test accuracy (**56.54%**), indicating it captured the temporal dependencies in the data more effectively.
   - This is expected because LSTM units are designed to retain long-term information using their memory cells and gating mechanisms (input, forget, and output gates).

2. **GRU Performed Moderately Well**  
   - GRU reached a reasonable accuracy of **53.29%**, which is slightly lower than LSTM but significantly better than Vanilla RNN.
   - GRUs are known to be computationally efficient and often perform comparably to LSTMs. In this case, their simpler structure (fewer gates) may have limited their ability to model complex patterns in the data.

3. **Vanilla RNN Showed Poor Performance**  
   - The lowest accuracy (**43.68%**) was observed in the Vanilla RNN model.
   - This architecture lacks gating mechanisms, making it prone to the **vanishing gradient problem**, which prevents effective learning over long sequences.

---

###  Interpretation:

- **Sequence Learning Capability**: The results reflect that the more advanced the recurrent architecture (in terms of memory handling), the better the model learns sequential patterns.
- **Trade-off**: LSTM offers higher accuracy at the cost of increased training time and complexity. GRU provides a balance between performance and computation.
- **Suitability**: 
  - If computational resources are limited, **GRU** can be a good choice.
  - For applications where accuracy is critical and computation is not a constraint, **LSTM** is preferable.
  - **Vanilla RNN** is not recommended for long sequences or tasks requiring memory retention.

---

###  Conclusion:

LSTM was the best performer, demonstrating the importance of using gated mechanisms for learning long-term dependencies in sequential data. GRU is a viable alternative with decent performance. Vanilla RNNs, although simpler, are less effective for complex sequence learning tasks.
